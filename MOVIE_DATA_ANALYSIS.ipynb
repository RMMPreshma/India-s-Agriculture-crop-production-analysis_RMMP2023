DATA CLEANING

Finding null values

```python
df.isnull()
```

```python
df.isnull().sum()#metascore,certficate,gross columns has null values
```

```python
# Remove commas and in gross column and convert  the data type of Gross column into float
#data formatting

df['Gross']=df['Gross'].str.replace(',','').astype(float)
```

```python
#clean missing values
df['Meta_score'].fillna(df['Meta_score'].mean(),inplace=True)
df['Certificate'].fillna('Not specified',inplace=True)
df['Gross'].fillna(df['Gross'].mean(),inplace=True)
```

Cleaning duplicates

```python
df.duplicated()
```

```python
df = df.drop_duplicates()
```

```python
#see the duplicated values in the dataset
df[df.duplicated()]
```

DATA MANIPULATION

```python
# Drop unnecessary columns
df = df.drop(['Poster_Link', 'Overview'], axis=1)
```

```python
df.shape
```

DATA ANALYSIS

```python
# Correlation between 'IMDB_Rating' and 'Gross'
correlation = df['IMDB_Rating'].corr(df['Gross'])
print(correlation)
```

```python
# Genre distribution
genre_distribution = df['Genre'].value_counts()
print(genre_distribution)
#More Movies released under drama genre
```

```python
# Calculate average gross revenue for each genre
avg_gross_genre = df.groupby('Genre')['Gross'].mean()
# Identify genre with highest average gross revenue
top_genre_gross = avg_gross_genre.idxmax()
top_genre_avg_gross = avg_gross_genre.max()
print(top_genre_gross)
print(top_genre_avg_gross)
#family,scifi movies  collected highest gross revenue
```

```python
# Director with highest average rating
director_avg_rating = df.groupby('Director')['IMDB_Rating'].mean()
best_director = director_avg_rating.idxmax()
print(best_director)
```

```python
# Calculate average IMDb rating for each year
avg_rating_yearly = df.groupby('Released_Year')['IMDB_Rating'].mean()
print(avg_rating_yearly)
```

```python
#  which  year has higherst rating?
max_rating_year = avg_rating_yearly.idxmax()
max_rating_value = avg_rating_yearly.max()
print(max_rating_year)
print(max_rating_value)
```

```python
# Calculate average Gross(Revenue) for each year
total_revenue_yearly = df.groupby('Released_Year')['Gross'].sum()
print(total_revenue_yearly)
```

```python
#  which  year has higherst gross?
max_rating_year = total_revenue_yearly.idxmax()
max_rating_value = total_revenue_yearly.max()
print(max_rating_year)
print(max_rating_value)
```
